```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Get files
from google.colab import files

print('Open excel file of Bank Statement!')
bank_stat_address = files.upload()

for keyb in bank_stat_address.keys():
  bank_file = str(keyb)
print(' ')
print('Open excel file of Company Bank movements!')
company_bank_address = files.upload()

for keyc in company_bank_address.keys():
  company_file = str(keyc)
```

```python
bank = pd.read_excel(bank_file)
company = pd.read_excel(company_file)

bank.set_index(bank.iloc[:,0], inplace=True)
company.set_index(company.iloc[:,0], inplace=True)

# bank.dropna(subset=['Statement Value Date'], inplace=True)
# company.dropna(subset=['Statement Value Date'], inplace=True)
bank = bank[bank.index.notnull()]  # remove nan values from first column (index row)
bank = bank.fillna(0) # replace nan values by zero
bank.loc['Total']= bank.sum(numeric_only=True) # Add last line for Total
bank.fillna('', inplace=True)


company = company[company.index.notnull()]
company = company.fillna(0) # replace nan values by zero
company.loc['Total']= company.sum(numeric_only=True) # Add last line for Total
company.fillna('', inplace=True) # remove NAN values from total line


print('*** Bank Statement ***')
print(bank)
print(' ')
print('*** Company Bank Movements ***')
print(company)
```
