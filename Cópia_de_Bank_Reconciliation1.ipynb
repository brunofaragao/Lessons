```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Get files
from google.colab import files

print('Open excel file of Bank Statement!')
bank_stat_address = files.upload()

for keyb in bank_stat_address.keys():
  bank_file = str(keyb)
print(' ')
print('Open excel file of Company Bank movements!')
company_bank_address = files.upload()

for keyc in company_bank_address.keys():
  company_file = str(keyc)
```

```python
bank = pd.read_excel(bank_file)  # Read file from excel
bank_date_col = bank.columns[0]  # get the name of the first column
bank.dropna(subset=[bank_date_col], inplace=True) # Remove NAN values from first column
bank = bank.fillna(0) # replace nan values by zero in the others columns
bank[bank_date_col] = bank[bank_date_col].astype('string') # Convert first column to string
bank['Sum'] = bank['Debit'].round(2)-bank['Credit'].round(2) # Add a Total row at the bottom
bank['Concat'] = bank[bank_date_col] + ' ' + bank['Sum'].astype(str)
# bank.set_index(bank.iloc[:,0], inplace=True) # Set the first column as index
bank = bank[bank.index.notnull()]  # remove NAN values from first column (index row)
bank.loc['Total']= bank.sum(numeric_only=True) # Add last line for Total
bank.fillna('', inplace=True) # Remove NAN values with a ''

company = pd.read_excel(company_file)
company_date_col = company.columns[0]  # get the name of the first column
company.dropna(subset=[company_date_col], inplace=True) # Remove NAN values from first column
company = company.fillna(0) # replace nan values by zero in the others columns
company[company_date_col] = company[company_date_col].astype('string') # Convert first column to string
company['Sum'] = company['Debit'].round(2)-company['Credit'].round(2) # Add a Total row at the bottom
company['Concat'] = company[company_date_col] + ' ' + company['Sum'].astype(str)
# company.set_index(company.iloc[:,0], inplace=True) # Set the first column as index
company = company[company.index.notnull()]  # remove NAN values from first column (index row)
company.loc['Total']= company.sum(numeric_only=True) # Add last line for Total
company.fillna('', inplace=True) # Remove NAN values with a ''


print('*** Bank Statement ***')
print(bank)
print(' ')
print('*** Company Bank Movements ***')
print(company)
```

```python
print('The totals are displayed under:')
print('-------------------------------')
print('Bank Statment')
print(' ')
print(bank.loc['Total'])

print('-------------------------------')
print('Company Report')
print(' ')
print(company.loc['Total'])
```

```python
Bank = bank[['Sum', 'Concat']]
Company = company[['Sum', 'Concat']]
print('*** Bank Statement ***')
print(Bank)
print(' ')
print('*** Company Bank Movements ***')
print(Company)
```

```python
form={'left_only':'Found in Bank, does not match with Company', 'right_only':'Found in Company, does not match with Bank', 'both':'Found in both, MATCH!'}
result = pd.merge(Bank, Company, how='outer', on='Concat', indicator='Status')
result['Status'] = result['Status'].map(form).astype(str)
result.fillna('', inplace=True)
result.to_excel('Reconciliation.xlsx')
print(result)
```
