```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Get files
from google.colab import files

print('Open excel file of Bank Statement!')
bank_stat_address = files.upload()

for keyb in bank_stat_address.keys():
  bank_file = str(keyb)
print(' ')
print('Open excel file of Company Bank movements!')
company_bank_address = files.upload()

for keyc in company_bank_address.keys():
  company_file = str(keyc)
```

```python
bank = pd.read_excel(bank_file)
company = pd.read_excel(company_file)

bank.set_index(bank.iloc[:,0], inplace=True)
company.set_index(company.iloc[:,0], inplace=True)

bank.dropna(subset=['Statement Value Date'], inplace=True)
company.dropna(subset=['Statement Value Date'], inplace=True)

print('*** Bank Statement ***')
print(bank)
print(' ')
print('*** Company Bank Movements ***')
print(company)
```
